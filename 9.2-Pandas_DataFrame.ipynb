```python
df.groupby("original_language").size()
```

```python
df.groupby("original_language")["vote_average"].value_counts()
```

```python
df.groupby("original_language")["vote_average"].agg(["count", "min", "max", "mean"])
```

```python
df[(df.original_language == "tr") & (df.vote_average > 7)].sort_values(["runtime"], ascending = False)
```

```python
df.dropna(subset = ["genre"], axis = 0, inplace = True)
```

```python
df[df.genre.str.contains("Documentary")]
```

```python
df["upper_title"] = df.original_title.str.upper()

df[["original_title", "upper_title"]]
```

```python
import string
def remove_punc(text):
    title = text.translate(str.maketrans('', '', string.punctuation))
    return title

df["remove_punc"] = df.original_title.apply(remove_punc)

df[["original_title", "upper_title", "remove_punc"]]
```
